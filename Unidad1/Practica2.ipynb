```python
from statistics import mean , median , mode , multimode
import random
```

```python
# lista de calificaciones del grupo de probabilidad y estatidistica
grupoa = [70, 70, 65, 90, 100, 95, 45, 80, 73, 20, 100, 96]
print(grupoa)
```

```python
# Sacar la media de las calificaciones del grupo
promedio = sum(grupoa) / len(grupoa)
print(promedio)
```

```python
mediana1 = median(grupoa) # Call the median function from statics module
print(mediana1)
```

```python
moda1 = mode(grupoa)
print(moda1)
```

```python
moda2 = multimode(grupoa)
print(moda2)
```

```python
fabrica1 = []
for i in range(100):
  numero_aleatorio = round(random.uniform(50,100), 2)
  fabrica1.append(numero_aleatorio)
print(fabrica1)
```

```python
# Libreria de python para graficar valores
import matplotlib.pyplot as plt
```

```python
plt.hist(fabrica1)
plt.title("Produccion diaria de la fabrica 1 ")
```

```python
media1 = mean(fabrica1)
mediana = median(fabrica1)
moda1 = mode(fabrica1)

print("valores de la fabrica1")
print("la media es: ", media1)
print("la mediana es: ", mediana1)
print("la moda es: ",moda1)
```
