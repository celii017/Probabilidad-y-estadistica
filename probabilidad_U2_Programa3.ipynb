```python
 import math as mt # Para utilizar factorial
 from itertools import permutations
 from itertools import product
 import numpy as np
```

```python
asistentes = ["a", "b", "c", "d", "f", "g", "h", "i", "j"]
n = len(asistentes)
r = 4 # Representa nuestro tama√±o de subconjuntos

pr = mt.factorial(n) / mt.factorial(n-r)
print("Existen ", pr, "posibles permutaciones sin repeticion del total de", n,
      " asistentes ", " en grupos de ", r, "en", r, "en", r)
print("Aqui las permutaciones")
permutacionesAsistentes = list(permutations(asistentes, r))

#Imprime las permutaciones
for permutacion in permutacionesAsistentes:
    print(permutacion)
```
